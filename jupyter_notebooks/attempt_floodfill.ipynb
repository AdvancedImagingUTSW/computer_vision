```python
from tifffile import imread, imsave
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, filters, color, morphology
from skimage.segmentation import flood, flood_fill
```

```python
im_data = imread('/home2/kdean/Desktop/minimum748447634.tif')
im_data = np.array(im_data)
im_data = np.int8(im_data)

print(np.shape(im_data))
```

```python
import pyclesperanto_prototype as cle

# Layer Result of gaussian_blur
image1 = cle.create_like(im_data)
image1 = cle.gaussian_blur(im_data, image1, 2.0, 2.0, 2.0)
print(np.shape(image1))

fig, ax = plt.subplots(ncols=2, figsize=(10, 5))
ax[0].imshow(im_data[224,:,:], cmap=plt.cm.gray)
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].imshow(image1[224,:,:], cmap=plt.cm.gray)
ax[1].plot(369, 308, 'ro')  # seed point
ax[1].set_title('Blurred')
ax[1].axis('off')

plt.show()
imsave('/home2/kdean/Desktop/output.tif',image1)
```
